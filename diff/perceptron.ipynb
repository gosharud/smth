```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def fit(x):
  return (x < 0)
```

```python
W1 = np.array([[1, -1, 0.5],
               [1, -1, -0.5],
               [1, 1, 0.5],
               [1, 1, -0.5]])
W2 = np.array([1, -1, 1, -1, 1])
```

```python
def perceptron(x):
    z1 = fit(W1 @ np.concatenate([x, np.ones(1)]).T)
    z2 = fit(W2 @ np.concatenate((z1, np.ones(1))).T)
    return z2
```

```python
np.random.seed(42)
X = np.random.random((2500, 2)) * 4 - 2
Y = np.array([perceptron(x) for x in X])
red = X[Y == 0]
green = X[Y == 1]
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(red[:, 0], red[:, 1], c='red')
plt.scatter(green[:, 0], green[:, 1], c='green')
plt.plot([-0.5, 0, 0.5, 0, -0.5], [0, 0.5, 0, -0.5, 0], c='blue')
plt.show()
```
